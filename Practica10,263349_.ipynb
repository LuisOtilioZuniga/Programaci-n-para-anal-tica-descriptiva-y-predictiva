```python
#Práctica #10: Análisis de Ventas con GridSpec
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec

# Lista de meses
meses = ["Enero", "Febrero", "Marzo", "Abril", "Mayo", "Junio",
         "Julio", "Agosto", "Septiembre", "Octubre", "Noviembre", "Diciembre"]

# Generar datos aleatorios
ventas = np.random.randint(2000, 5000, size=12)   # entre 2000 y 5000
gastos = np.random.randint(1000, 4000, size=12)   # entre 1000 y 4000

# Crear DataFrame
df = pd.DataFrame({
    "Mes": meses,
    "Ventas": ventas,
    "Gastos": gastos
})

print(df)
```

```python
fig = plt.figure(figsize=(12, 6))
gs = gridspec.GridSpec(nrows=2, ncols=3, figure=fig)

# 1) Barra
ax_bar = fig.add_subplot(gs[0, 0:2])
ax_bar.bar(df["Mes"], df["Ventas"], edgecolor="black", color="skyblue")
ax_bar.set_title("Ventas por Mes (Barras)")
ax_bar.tick_params(axis="x", rotation=45)

# 2) Línea
ax_line = fig.add_subplot(gs[1, 0])
ax_line.plot(df["Mes"], df["Gastos"], marker="o", color="green")
ax_line.set_title("Gastos por Mes (Línea)")
ax_line.tick_params(axis="x", rotation=45)

# 3) Dispersión
ax_scatter = fig.add_subplot(gs[:, 2])
ax_scatter.scatter(df["Ventas"], df["Gastos"], color="red")
ax_scatter.set_title("Relación Ventas vs Gastos")
ax_scatter.set_xlabel("Ventas")
ax_scatter.set_ylabel("Gastos")

# Título general de la figura
fig.suptitle("Análisis de Ventas y Gastos", fontsize=14, fontweight="bold")

# Ajustar espacios
fig.tight_layout(rect=[0, 0, 1, 0.95])  # deja espacio arriba para el título general

plt.show()
```
