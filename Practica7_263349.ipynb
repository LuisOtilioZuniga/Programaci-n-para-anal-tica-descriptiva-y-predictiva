```python
# PRÁCTICA 7

import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

# Cargar dataset Iris desde seaborn
df = sns.load_dataset('iris')

# Calcular medias agrupadas por especie
means = df.groupby("species")[["petal_length", "petal_width"]].mean()

# Vista previa
print("Vista previa del dataset Iris:")
print(df.head())

# Información general
print("\nInformación general del dataset:")
print(df.info())

# Resumen estadístico
print("\nResumen estadístico de las características:")
print(df.describe())
```

```python
#GRÁFICO DE BARRAS

means.plot(kind="bar", figsize=(8,5))

plt.title("Promedio de largo y ancho de pétalos por especie")
plt.ylabel("Medida en cm")
plt.xlabel("Especie")
plt.xticks(rotation=0)  # Mantener nombres de especies horizontales
plt.legend(title="Características")
plt.show()
```

```python
# Crear histograma de sepal_length
plt.figure(figsize=(8,5))
plt.hist(df['sepal_length'], bins=15, color='skyblue', edgecolor='black')

# Personalización
plt.title("Distribución de la longitud de sépalos", fontsize=14)
plt.xlabel("Longitud del sépalo (cm)", fontsize=12)
plt.ylabel("Frecuencia", fontsize=12)

plt.show()
```

```python
# Crear gráfico de dispersión
plt.figure(figsize=(8,6))
sns.scatterplot(
    data=df,
    x="petal_length",
    y="petal_width",
    hue="species",    # color por especie
    palette="Set1",   # paleta de colores
    s=60              # tamaño de los puntos
)

# Personalización
plt.title("Relación entre largo y ancho de pétalos", fontsize=14)
plt.xlabel("Largo del pétalo (cm)", fontsize=12)
plt.ylabel("Ancho del pétalo (cm)", fontsize=12)

plt.legend(title="Especie")
plt.show()
```

```python
# Crear gráfico de cajas para sepal_length y sepal_width
plt.figure(figsize=(12,5))

# Boxplot para sepal_length
plt.subplot(1,2,1)  # 1 fila, 2 columnas, primer gráfico
sns.boxplot(data=df, x="species", y="sepal_length", palette="Set2")
plt.title("Distribución del largo del sépalo por especie")
plt.xlabel("Especie")
plt.ylabel("Longitud del sépalo (cm)")

# Boxplot para sepal_width
plt.subplot(1,2,2)  # 1 fila, 2 columnas, segundo gráfico
sns.boxplot(data=df, x="species", y="sepal_width", palette="Set2")
plt.title("Distribución del ancho del sépalo por especie")
plt.xlabel("Especie")
plt.ylabel("Ancho del sépalo (cm)")

plt.tight_layout()
plt.show()
```

```python
# Crear gráfico de líneas
plt.figure(figsize=(10,5))
plt.plot(df.index, df['petal_length'], color='purple', marker='o', linestyle='-', markersize=4)

# Personalización
plt.title("Tendencia de la longitud del pétalo a lo largo de las observaciones")
plt.xlabel("Índice de la observación")
plt.ylabel("Largo del pétalo (cm)")

plt.grid(True)
plt.show()
```
