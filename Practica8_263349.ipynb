```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

# Cargar dataset tips
df = sns.load_dataset('tips')

# Vista previa de los primeros registros
print("Primeras filas del dataset tips:")
print(df.head())

# Información general del dataset
print("\nInformación general del dataset:")
print(df.info())

# Resumen estadístico de las columnas numéricas
print("\nResumen estadístico de las columnas numéricas:")
print(df.describe())

# Verificar valores nulos
print("\nCantidad de valores nulos por columna:")
print(df.isnull().sum())
```

```python
# Calcular la matriz de correlación
corr_matrix = df[['total_bill', 'tip', 'size']].corr()

# Crear el mapa de calor
plt.figure(figsize=(6,5))
sns.heatmap(
    corr_matrix,
    annot=True,        # mostrar los valores de correlación
    cmap='coolwarm',   # paleta de colores
    fmt=".2f",         # formato de los números
    linewidths=0.5,
    square=True
)

plt.title("Mapa de calor de correlación - variables numéricas", fontsize=14)
plt.show()
```

```python
# Crear diagrama de violín
plt.figure(figsize=(8,5))
sns.violinplot(
    data=df,
    x="day",
    y="tip",
    palette="Set2"
)

# Personalización
plt.title("Distribución de propinas por día de la semana", fontsize=14)
plt.xlabel("Día de la semana", fontsize=12)
plt.ylabel("Propina ($)", fontsize=12)

plt.show()
```

```python
# Crear gráfico de dispersión
plt.figure(figsize=(8,6))
sns.scatterplot(
    data=df,
    x="total_bill",
    y="tip",
    hue="day",       # color por día
    palette="Set1",
    s=60             # tamaño de los puntos
)

# Personalización
plt.title("Relación entre total de la cuenta y propina por día", fontsize=14)
plt.xlabel("Total de la cuenta ($)", fontsize=12)
plt.ylabel("Propina ($)", fontsize=12)
plt.legend(title="Día de la semana")
plt.grid(True)
plt.show()
```

```python
# Calcular promedio de propinas por género
avg_tip_sex = df.groupby("sex")["tip"].mean().reset_index()

print("Promedio de propinas por género:")
print(avg_tip_sex)

# Crear gráfico de barras
plt.figure(figsize=(6,5))
sns.barplot(
    data=avg_tip_sex,
    x="sex",
    y="tip",
    palette="Pastel1"
)

# Personalización
plt.title("Promedio de propina por género", fontsize=14)
plt.xlabel("Género", fontsize=12)
plt.ylabel("Promedio de propina ($)", fontsize=12)

plt.show()
```
