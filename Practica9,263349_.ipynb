```python
#PRÁCTICA 9: DISTRIBUCION DE DATOS CON plt.subplots()
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Generar los 1000 valores para cada distribución
normal = np.random.normal(loc=0, scale=1, size=1000)       # media=0, desviación estándar=1
uniforme = np.random.uniform(low=0, high=1, size=1000)     # entre 0 y 1
exponencial = np.random.exponential(scale=1, size=1000)    # lambda=1
poisson = np.random.poisson(lam=3, size=1000)              # media=3

# Crear el DataFrame
df = pd.DataFrame({
    "Normal": normal,
    "Uniforme": uniforme,
    "Exponencial": exponencial,
    "Poisson": poisson
})

print(df.head())  # mostrar las primeras filas
```

```python
# Crear la figura y los ejes en formato 2x2
fig, ax = plt.subplots(2, 2, figsize=(10, 8))

# Graficar cada distribución en su respectivo subplot
ax[0, 0].hist(df["Normal"], bins=30, color="skyblue", edgecolor="black")
ax[0, 0].set_title("Distribución Normal")

ax[0, 1].hist(df["Uniforme"], bins=30, color="lightgreen", edgecolor="black")
ax[0, 1].set_title("Distribución Uniforme")

ax[1, 0].hist(df["Exponencial"], bins=30, color="salmon", edgecolor="black")
ax[1, 0].set_title("Distribución Exponencial")

ax[1, 1].hist(df["Poisson"], bins=30, color="orange", edgecolor="black")
ax[1, 1].set_title("Distribución Poisson")

# Ajustar el espacio entre los subplots
plt.tight_layout()

# Mostrar la figura
plt.show()
```
