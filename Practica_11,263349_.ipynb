```python
#Pr치ctica #11: Seaborn y Subplots
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

# Cargar dataset penguins
df = sns.load_dataset("penguins")

# Mostrar las primeras filas
print(df.head())
```

```python
# Crear figura con 2 filas y 2 columnas
fig, axs = plt.subplots(2, 2, figsize=(10, 8))

# (0,0) Histograma de la longitud de aletas
sns.histplot(data=df, x="flipper_length_mm", kde=True, color="skyblue", ax=axs[0,0])
axs[0,0].set_title("Histograma - Longitud de Aletas")

# (0,1) Boxplot masa corporal por especie
sns.boxplot(data=df, x="species", y="body_mass_g", palette="Set2", ax=axs[0,1])
axs[0,1].set_title("Boxplot - Masa Corporal por Especie")

# (1,0) Dispersi칩n masa corporal vs longitud de aletas
sns.scatterplot(data=df, x="body_mass_g", y="flipper_length_mm", hue="species", palette="Set1", ax=axs[1,0])
axs[1,0].set_title("Dispersi칩n - Masa Corporal vs Longitud de Aletas")

# (1,1) Violin plot masa corporal por especie
sns.violinplot(data=df, x="species", y="body_mass_g", palette="pastel", ax=axs[1,1])
axs[1,1].set_title("Violin Plot - Masa Corporal por Especie")

# Ajustar dise침o
plt.tight_layout()
plt.show()
```
