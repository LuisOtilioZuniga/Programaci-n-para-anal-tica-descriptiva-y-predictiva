# **Ejercicio 1: Normalización Min-Max**

1.	Cargar el dataset wine de sklearn.
2.	Seleccionar una de las características numéricas (por ejemplo, alcohol).
3.	Realizar una prueba de normalidad (usando scipy.stats.shapiro) sobre los datos originales.
4.	Aplicar la normalización Min-Max a los datos seleccionados.
5.	Realizar la prueba de normalidad nuevamente sobre los datos normalizados.
6.	Graficar los datos originales y los normalizados utilizando matplotlib o seaborn.

2. ¿Qué interpretas del p-valor antes y después de la normalización?


```python
# Ejercicio 1: Normalización Min-Max

# Importar las librerías necesarias
from sklearn.datasets import load_wine         # Para cargar el dataset wine
from sklearn.preprocessing import MinMaxScaler # Para aplicar la normalización Min-Max
from scipy.stats import shapiro                # Para realizar la prueba de normalidad
import matplotlib.pyplot as plt                # Para graficar
import seaborn as sns                          # Para mejorar el estilo de las gráficas
import pandas as pd                            # Para manejar los datos más fácilmente

# 1. Cargar el dataset wine
wine = load_wine()  # Cargamos el dataset directamente desde sklearn

# Convertimos los datos a un DataFrame para trabajar más cómodo
df = pd.DataFrame(wine.data, columns=wine.feature_names)

# 2. Seleccionar una característica numérica (ejemplo: alcohol)
alcohol = df['alcohol']  # Seleccionamos la columna "alcohol"

# 3. Prueba de normalidad sobre los datos originales
stat_original, p_original = shapiro(alcohol)
print("=== Prueba de normalidad (Datos Originales) ===")
print(f"Estadístico W: {stat_original:.4f}")
print(f"P-valor: {p_original:.4f}")

# 4. Aplicar la normalización Min-Max
scaler = MinMaxScaler()  # Creamos el objeto de normalización
alcohol_normalizado = scaler.fit_transform(alcohol.values.reshape(-1, 1))  # Ajustamos y transformamos

# 5. Prueba de normalidad sobre los datos normalizados
stat_norm, p_norm = shapiro(alcohol_normalizado)
print("\n=== Prueba de normalidad (Datos Normalizados) ===")
print(f"Estadístico W: {stat_norm:.4f}")
print(f"P-valor: {p_norm:.4f}")

# 6. Graficar los datos originales y normalizados
plt.figure(figsize=(12, 5))

# Gráfico de los datos originales
plt.subplot(1, 2, 1)
sns.histplot(alcohol, kde=True, color='blue')
plt.title("Distribución de Alcohol (Original)")
plt.xlabel("Alcohol")
plt.ylabel("Frecuencia")

# Gráfico de los datos normalizados
plt.subplot(1, 2, 2)
sns.histplot(alcohol_normalizado.flatten(), kde=True, color='green')
plt.title("Distribución de Alcohol (Normalizado Min-Max)")
plt.xlabel("Alcohol Normalizado")
plt.ylabel("Frecuencia")

plt.tight_layout()
plt.show()
```

**Conclusiones: Ejercicio 1: Normalización Min-Max**

¿Cómo afectó la normalización Min-Max a la distribución de los datos?
La normalización Min-Max no cambia la forma de la distribución, solo ajusta el rango de los valores entre 0 y 1. Es decir, si los datos originales estaban sesgados o no eran normales, seguirán teniendo una forma similar, solo que ahora todos los valores están dentro del rango [0, 1].

¿Qué interpretas del p-valor antes y después de la normalización?
El p-valor nos indica si los datos siguen una distribución normal.

- Si p < 0.05, los datos no son normales.
- Si p ≥ 0.05, no se puede rechazar la normalidad.

Después de la normalización, el p-valor puede cambiar ligeramente, ya que los valores están reescalados, pero usualmente la conclusión de normalidad no cambia significativamente.

# **Ejercicio 2: Normalización Z-Score**

Instrucciones:

1.	Cargar el dataset wine de sklearn.
2.	Seleccionar una de las características numéricas (por ejemplo, malic_acid).
3.	Realizar una prueba de normalidad (usando scipy.stats.shapiro) sobre los datos originales.
4.	Aplicar la normalización Z-Score a los datos seleccionados.
5.	Realizar la prueba de normalidad nuevamente sobre los datos normalizados.
6.	Graficar los datos originales y los normalizados utilizando matplotlib o seaborn.

```python
# Ejercicio 2: Normalización Z-Score

# Importar librerías necesarias
from sklearn.datasets import load_wine         # Para cargar el dataset wine
from sklearn.preprocessing import StandardScaler  # Para aplicar la normalización Z-Score
from scipy.stats import shapiro                # Para la prueba de normalidad
import matplotlib.pyplot as plt                # Para las gráficas
import seaborn as sns                          # Para mejorar el estilo visual
import pandas as pd                            # Para manejar los datos fácilmente

# 1. Cargar el dataset wine
wine = load_wine()  # Cargamos el dataset directamente desde sklearn

# Convertimos los datos en un DataFrame
df = pd.DataFrame(wine.data, columns=wine.feature_names)

# 2. Seleccionar una característica numérica (ejemplo: malic_acid)
malic_acid = df['malic_acid']  # Seleccionamos la columna "malic_acid"

# 3. Prueba de normalidad sobre los datos originales
stat_original, p_original = shapiro(malic_acid)
print("=== Prueba de normalidad (Datos Originales) ===")
print(f"Estadístico W: {stat_original:.4f}")
print(f"P-valor: {p_original:.4f}")

# 4. Aplicar la normalización Z-Score
scaler = StandardScaler()  # Creamos el objeto para la normalización Z-Score
malic_acid_normalizado = scaler.fit_transform(malic_acid.values.reshape(-1, 1))  # Aplicamos la normalización

# 5. Prueba de normalidad sobre los datos normalizados
stat_norm, p_norm = shapiro(malic_acid_normalizado)
print("\n=== Prueba de normalidad (Datos Normalizados Z-Score) ===")
print(f"Estadístico W: {stat_norm:.4f}")
print(f"P-valor: {p_norm:.4f}")

# 6. Graficar los datos originales y los normalizados
plt.figure(figsize=(12, 5))

# Gráfico de los datos originales
plt.subplot(1, 2, 1)
sns.histplot(malic_acid, kde=True, color='orange')
plt.title("Distribución de Malic Acid (Original)")
plt.xlabel("Malic Acid")
plt.ylabel("Frecuencia")

# Gráfico de los datos normalizados
plt.subplot(1, 2, 2)
sns.histplot(malic_acid_normalizado.flatten(), kde=True, color='green')
plt.title("Distribución de Malic Acid (Normalizado Z-Score)")
plt.xlabel("Malic Acid Normalizado")
plt.ylabel("Frecuencia")

plt.tight_layout()
plt.show()
```

**Conclusiones: Ejercicio 2: Normalización Z-Score**

¿Qué diferencia observas entre la distribución original y la normalizada con Z-Score?
La distribución Z-Score mantiene la misma forma que la original, pero ahora está centrada alrededor de cero (media = 0) y con una desviación estándar igual a 1.
Esto significa que los valores ahora se miden en función de cuántas desviaciones estándar se alejan de la media, lo que permite comparar distintas variables sin importar sus unidades o escalas.


¿Qué implicaciones tiene la normalización Z-Score en el análisis de los datos?
La normalización Z-Score es muy útil en técnicas estadísticas y de machine learning, ya que:

- Evita que variables con escalas grandes dominen a las demás.
- Facilita la convergencia en algoritmos que usan distancias (como KNN, SVM o PCA).
- Permite comparar variables distintas en igualdad de condiciones.

En resumen, la normalización Z-Score homogeneiza la escala de los datos, manteniendo la forma de su distribución original.


# **Ejercicio 3: Transformación Logarítmica**

Instrucciones:

1.	Cargar un conjunto de datos aleatorios generados con numpy.
2.	Aplicar una transformación logarítmica (np.log) a los datos.
3.	Realizar una prueba de normalidad sobre los datos originales y transformados.
4.	Graficar los datos originales y transformados utilizando matplotlib.
5.	Responder las preguntas siguientes.

Preguntas:
1.	¿Cómo afectó la transformación logarítmica a la distribución de los datos?
2.	¿Qué interpretas del p-valor antes y después de la transformación logarítmica?


```python
# Ejercicio 3: Transformación Logarítmica

# Importar las librerías necesarias
import numpy as np                   # Para generar datos aleatorios y aplicar logaritmos
from scipy.stats import shapiro      # Para la prueba de normalidad
import matplotlib.pyplot as plt      # Para graficar los datos
import seaborn as sns                # Para darle mejor estilo a las gráficas

# 1. Generar un conjunto de datos aleatorios
# Generaremos datos con distribución exponencial (que no es normal)
# para poder observar cómo la transformación logarítmica mejora su simetría
np.random.seed(42)  # Fijamos semilla para reproducibilidad
datos_originales = np.random.exponential(scale=2.0, size=200)  # 200 datos aleatorios

# 2. Aplicar la transformación logarítmica
# np.log transforma los datos positivos reduciendo la asimetría a la derecha
datos_log = np.log(datos_originales)

# 3. Prueba de normalidad (Shapiro-Wilk)
stat_original, p_original = shapiro(datos_originales)
stat_log, p_log = shapiro(datos_log)

print("=== Prueba de Normalidad ===")
print("Datos Originales:")
print(f"Estadístico W = {stat_original:.4f}, P-valor = {p_original:.4f}")
print("\nDatos Transformados (Log):")
print(f"Estadístico W = {stat_log:.4f}, P-valor = {p_log:.4f}")

# 4. Graficar los datos originales y transformados
plt.figure(figsize=(12, 5))

# Histograma de los datos originales
plt.subplot(1, 2, 1)
sns.histplot(datos_originales, kde=True, color='red')
plt.title("Distribución Original (Exponencial)")
plt.xlabel("Valores originales")
plt.ylabel("Frecuencia")

# Histograma de los datos transformados
plt.subplot(1, 2, 2)
sns.histplot(datos_log, kde=True, color='green')
plt.title("Distribución Transformada (Logarítmica)")
plt.xlabel("Valores logarítmicos")
plt.ylabel("Frecuencia")

plt.tight_layout()
plt.show()
```

**Conclusiones: Ejercicio 3: Transformación Logarítmica Instrucciones:**

¿Cómo afectó la transformación logarítmica a la distribución de los datos?
La transformación logarítmica reduce la asimetría (sesgo) hacia la derecha, haciendo que los datos se acerquen más a una distribución simétrica o normal.
En otras palabras, los valores grandes se comprimen y los pequeños se expanden, suavizando la forma de la distribución.

¿Qué interpretas del p-valor antes y después de la transformación logarítmica?

- Si el p-valor aumenta después de la transformación, significa que los datos se acercaron a la normalidad.

- Si el p-valor < 0.05, aún se considera que los datos no son normales, pero la mejora puede ser evidente visualmente.

La transformación logarítmica suele ser útil cuando los datos presentan sesgo positivo (cola hacia la derecha), como los tiempos, precios o cantidades que no pueden ser negativas.

# **Ejercicio 4: Transformación Raíz Cuadrada**

Instrucciones:

1.	Cargar un conjunto de datos aleatorios generados con numpy.
2.	Aplicar una transformación de raíz cuadrada a los datos.
3.	Realizar una prueba de normalidad sobre los datos originales y transformados.
4.	Graficar los datos originales y transformados utilizando matplotlib.
5.	Responder las preguntas siguientes.

Preguntas:
1.	¿Cómo afectó la transformación raíz cuadrada a la distribución de los datos?
2.	¿Qué efectos tiene la raíz cuadrada sobre los valores extremos de los datos?


```python
# Ejercicio 4: Transformación Raíz Cuadrada
# Importar librerías necesarias
import numpy as np                   # Para generar datos aleatorios y aplicar raíz cuadrada
from scipy.stats import shapiro      # Para la prueba de normalidad
import matplotlib.pyplot as plt      # Para graficar
import seaborn as sns                # Para un mejor estilo visual

# 1. Generar un conjunto de datos aleatorios
# En este caso, generaremos datos con distribución exponencial,
# que suelen estar sesgados positivamente (cola hacia la derecha).
np.random.seed(42)  # Fijamos una semilla para reproducibilidad
datos_originales = np.random.exponential(scale=2.0, size=200)  # 200 valores positivos

# 2. Aplicar la transformación de raíz cuadrada
# La raíz cuadrada (np.sqrt) reduce la variabilidad de los valores grandes,
# haciendo que los datos sean más simétricos.
datos_raiz = np.sqrt(datos_originales)

# 3. Prueba de normalidad (Shapiro-Wilk)
stat_original, p_original = shapiro(datos_originales)
stat_raiz, p_raiz = shapiro(datos_raiz)

print("=== Prueba de Normalidad ===")
print("Datos Originales:")
print(f"Estadístico W = {stat_original:.4f}, P-valor = {p_original:.4f}")
print("\nDatos Transformados (Raíz Cuadrada):")
print(f"Estadístico W = {stat_raiz:.4f}, P-valor = {p_raiz:.4f}")

# 4. Graficar los datos originales y transformados
plt.figure(figsize=(12, 5))

# Gráfico de los datos originales
plt.subplot(1, 2, 1)
sns.histplot(datos_originales, kde=True, color='purple')
plt.title("Distribución Original (Exponencial)")
plt.xlabel("Valores originales")
plt.ylabel("Frecuencia")

# Gráfico de los datos transformados
plt.subplot(1, 2, 2)
sns.histplot(datos_raiz, kde=True, color='green')
plt.title("Distribución Transformada (Raíz Cuadrada)")
plt.xlabel("Valores transformados (√x)")
plt.ylabel("Frecuencia")

plt.tight_layout()
plt.show()
```

**Conclusiones: Ejercicio 4: Transformación Raíz Cuadrada**

¿Cómo afectó la transformación raíz cuadrada a la distribución de los datos?
La transformación de raíz cuadrada reduce la asimetría positiva (cola hacia la derecha) y hace que los datos sean más simétricos. Aunque no convierte los datos en perfectamente normales, acerca su forma a una distribución normal, especialmente cuando los datos originales están sesgados positivamente.

¿Qué efectos tiene la raíz cuadrada sobre los valores extremos de los datos?
La raíz cuadrada disminuye la magnitud de los valores grandes y suaviza la influencia de los outliers.
Esto ayuda a reducir la dispersión y estabilizar la varianza, haciendo que los datos sean más adecuados para análisis estadísticos o modelos sensibles a valores extremos.

# **Ejercicio 5: Transformación Box-Cox**

Instrucciones:

1.	Cargar un conjunto de datos aleatorios generados con numpy.
2.	Aplicar la transformación Box-Cox a los datos utilizando scipy.stats.boxcox.
3.	Realizar una prueba de normalidad sobre los datos originales y transformados.
4.	Graficar los datos originales y transformados utilizando matplotlib.
5.	Responder las preguntas siguientes.

Preguntas:
1.	¿Cómo afectó la transformación Box-Cox a la distribución de los datos?
2.	¿Qué interpretación tiene el valor lambda obtenido en la transformación Box-Cox?


```python
# Ejercicio 5: Transformación Box-Cox
# Importar librerías necesarias
import numpy as np                     # Para generar los datos aleatorios
from scipy.stats import shapiro, boxcox  # Para prueba de normalidad y transformación Box-Cox
import matplotlib.pyplot as plt         # Para graficar los resultados
import seaborn as sns                   # Para un mejor estilo visual

# 1. Generar un conjunto de datos aleatorios
# Generaremos datos con una distribución exponencial (positivos y sesgados a la derecha)
# porque la transformación Box-Cox solo funciona con valores positivos.
np.random.seed(42)  # Semilla para reproducibilidad
datos_originales = np.random.exponential(scale=2.0, size=200)

# 2. Aplicar la transformación Box-Cox
# La función boxcox devuelve los datos transformados y el valor lambda utilizado.
# Box-Cox busca automáticamente el mejor valor de lambda que acerque los datos a la normalidad.
datos_boxcox, lambda_boxcox = boxcox(datos_originales)

# 3. Prueba de normalidad (Shapiro-Wilk)
stat_original, p_original = shapiro(datos_originales)
stat_boxcox, p_boxcox = shapiro(datos_boxcox)

print("=== Prueba de Normalidad ===")
print("Datos Originales:")
print(f"Estadístico W = {stat_original:.4f}, P-valor = {p_original:.4f}")
print("\nDatos Transformados (Box-Cox):")
print(f"Estadístico W = {stat_boxcox:.4f}, P-valor = {p_boxcox:.4f}")
print(f"\nValor de Lambda encontrado: λ = {lambda_boxcox:.4f}")

# 4. Graficar los datos originales y transformados
plt.figure(figsize=(12, 5))

# Histograma de los datos originales
plt.subplot(1, 2, 1)
sns.histplot(datos_originales, kde=True, color='red')
plt.title("Distribución Original (Exponencial)")
plt.xlabel("Valores originales")
plt.ylabel("Frecuencia")

# Histograma de los datos transformados
plt.subplot(1, 2, 2)
sns.histplot(datos_boxcox, kde=True, color='green')
plt.title("Distribución Transformada (Box-Cox)")
plt.xlabel("Valores transformados")
plt.ylabel("Frecuencia")

plt.tight_layout()
plt.show()
```

**Conclusiones: Ejercicio 5: Transformación Box-Cox**

¿Cómo afectó la transformación Box-Cox a la distribución de los datos?
La transformación Box-Cox modificó los datos para que su distribución fuera más simétrica y cercana a la normal.
A diferencia de otras transformaciones (como log o raíz cuadrada), Box-Cox ajusta automáticamente la potencia (λ) que mejor corrige la asimetría.
Por tanto, mejora significativamente la normalidad de los datos cuando estos son positivos y sesgados.


¿Qué interpretación tiene el valor lambda obtenido en la transformación Box-Cox?
El parámetro λ (lambda) indica la potencia óptima aplicada a los datos para lograr una distribución más normal:

- Si λ = 1, la transformación apenas altera los datos (son casi normales).
- Si λ = 0, equivale a una transformación logarítmica.
- Si λ < 1, los valores grandes se reducen más, suavizando la cola derecha.

En resumen, λ controla el grado de corrección de asimetría y determina cómo los datos se reescalan para aproximarse a una forma normal.
