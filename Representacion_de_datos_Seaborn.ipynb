```python
import seaborn as sns
import matplotlib.pyplot as plt

titanic = sns.load_dataset('titanic')
print(titanic)
```

```python
sns.histplot(titanic['age'],kde=True, color = 'blue', bins = 20)
plt.title('Distribucion de Edades de los Pasajeros')
plt.xlabel('Edad')
plt.ylabel('Frecuencia')
plt.show()
```

```python
sns.barplot(x='sex', y = 'survived', data = titanic, palette = 'pastel')
plt.title('Tasa de Supervivencia')
plt.xlabel('Genero')
plt.ylabel('Tasa de Supervivencia')
plt.show()
```

```python
sns.catplot(x = 'pclass', hue = 'survived', kind = 'count', data = titanic, palette = 'Set2')
plt.title('Supervivencia por clase')
plt.xlabel('Clase')
plt.ylabel('Cantidad de Pasajeros')
plt.show()
```

```python
sns.boxplot(x = 'pclass', y = 'age', data = titanic, palette = 'coolwarm')
plt.title('Distribucion de Edades por Clase de Pasajeros')
plt.xlabel('Clase')
plt.ylabel('Edad')
plt.show()
```

```python
num_col = titanic.select_dtypes(include = ['number']).columns
corr = titanic[num_col].corr()
sns.heatmap(corr,annot = True, cmap = 'coolwarm', fmt = '.2f', linewidths = 0.5)
plt.title('Matriz de Correlacion')
plt.show()
```

```python
sns.pairplot(titanic[['age','fare','pclass','survived']])
plt.suptitle('Pairplot de las variables de Titanic')
plt.show()
```

```python
sns.violinplot(x = 'pclass', y = 'age', data = titanic, palette = 'muted')
plt.title('Distribucion de Edades de los Pasajeros')
plt.xlabel('Edad')
plt.ylabel('Frecuencia')
plt.show()
```

```python
sns.scatterplot(x = 'age', y= 'fare', data = titanic, hue = 'survived', palette = 'coolwarm')
plt.title('Relacion entre edad y tarifa de supervivencia')
plt.xlabel('Edad')
plt.ylabel('Clase')
plt.show()
```
